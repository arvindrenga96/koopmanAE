```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# These are the parameters for the Lorenz Attractor
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# This is the system of differential equations known as the Lorenz Attractor
def lorenz_system(current_state, t):
    x, y, z = current_state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]

# Initial state
initial_state = [1.0, 1.0, 1.0]

# Time points
t = np.arange(0.0, 50.0, 0.01)

# Solve differential equations
solution = odeint(lorenz_system, initial_state, t)

# Plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(solution[:, 0], solution[:, 1], solution[:, 2])
plt.title('Lorenz Attractor')
plt.show()
```

```python
# Plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(solution[:2000, 0], solution[:2000, 1], solution[:2000, 2])
plt.title('Lorenz Attractor Training')
plt.show()
```
